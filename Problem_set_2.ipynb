```python
def main():
#create sorted list
  sorted = [1,2,3,4,5,6,7,8,9]
# i is the beginnning of list  
  i = 0
# j is one placement above i 
  j = 1
# While j is less than the length of sorted list print true
  while j < len(sorted):
    print(True)
#If the sorted list element j is greater than sorted list element j - 1 end 
    if sorted[j] < sorted[j-1]:
      break
#increment both i and j 
    i += 1 
    j += 1

if __name__ == "__main__":
    main()
```
